```python
# Gun detection scene graph generator for office dataset creation
import json
import random

people_attrs = [["employee", "standing"], ["intruder", "agitated"], ["security_guard", "walking"], ["visitor", "sitting"], ["manager", "crouching"]]
weapons = [["handgun", "raised"], ["pistol", "concealed"], ["assault_rifle", "lowered"], ["revolver", "on_table"], ["shotgun", "brandishing"]]
benign_objects = [
    ({"label": "Furniture", "attrs": ["desk"]}, {"label": "Electronics", "attrs": ["laptop"]}),
    ({"label": "Decor", "attrs": ["potted_plant"]}, {"label": "Furniture", "attrs": ["filing_cabinet"]}),
    ({"label": "Electronics", "attrs": ["printer"]}, {"label": "Object", "attrs": ["coffee_mug"]}),
    ({"label": "Furniture", "attrs": ["conference_table"]}, {"label": "Electronics", "attrs": ["projector"]}),
    ({"label": "Architecture", "attrs": ["glass_door"]}, {"label": "Decor", "attrs": ["water_cooler"]})
]
relationships = ["holding", "aiming", "concealing", "dropping", "reaching_for"]

def generate_scene_graphs(count, filename="gun_detection_scenes.jsonl"):
    with open(filename, 'w') as f:
        for i in range(1, count + 1):
            person = random.choice(people_attrs)
            weapon = random.choice(weapons)
            b_obj1, b_obj2 = random.choice(benign_objects)
            rel = random.choice(relationships)
            
            scene = {
                "scene_id": i,
                "environment": "Office",
                "event": "Gun Detection",
                "nodes": [
                    {"id": "n1", "label": "Person", "attributes": person},
                    {"id": "n2", "label": "Weapon", "attributes": weapon},
                    {"id": "n3", "label": b_obj1["label"], "attributes": b_obj1["attrs"]},
                    {"id": "n4", "label": b_obj2["label"], "attributes": b_obj2["attrs"]}
                ],
                "edges": [
                    {"source": "n1", "target": "n2", "relationship": rel},
                    {"source": "n1", "target": "n3", "relationship": "near"},
                    {"source": "n3", "target": "n4", "relationship": "adjacent_to"}
                ]
            }
            f.write(json.dumps(scene) + '\n')

generate_scene_graphs(500)
print("Successfully generated 500 scene graphs in 'gun_detection_scenes.jsonl'")
```

```python
import json
import random

# ==========================================
# 1. VOCABULARY MATCHING SGG OUTPUT
# ==========================================
ENVIRONMENTS = ["Home", "Classroom", "Office"]

SAFE_OBJECTS = ["Person", "Desk", "Chair", "Table", "Couch", "Monitor", "Laptop", "Backpack", "Book", "Cup", "Plant"]
THREAT_MAPPING = {
    "Gun Detection": ["Gun"],
    "Knife Detection": ["Knife"],
    "Fire Detection": ["Fire", "Smoke", "Flames", "Burning"],
    "Fall Detection": ["Person Falling"]
}

# The exact relationships generated by our Semantic Heuristics Engine
SAFE_EDGES = [
    "physically_interacting_with", 
    "sitting_on_or_standing_by", 
    "adjacent_to", 
    "conversing_or_facing"
]

def create_base_graph(environment, num_background_objects=5):
    """Creates the standard nodes and safe background edges."""
    nodes = []
    
    # Always include at least one normal Person in every scene
    nodes.append({"id": "n1", "label": "Person", "attributes": []})
    
    # Add random background objects
    for i in range(num_background_objects):
        obj = random.choice(SAFE_OBJECTS)
        nodes.append({"id": f"n{len(nodes)+1}", "label": obj, "attributes": []})
        
    edges = []
    # Generate a few safe relationships between random nodes
    for _ in range(random.randint(2, 5)):
        source_idx = random.randint(0, len(nodes)-1)
        target_idx = random.randint(0, len(nodes)-1)
        if source_idx != target_idx:
            edges.append({
                "source": nodes[source_idx]["id"],
                "target": nodes[target_idx]["id"],
                "relationship": random.choice(SAFE_EDGES)
            })
            
    return nodes, edges

# ==========================================
# 2. GENERATION LOGIC
# ==========================================
def generate_safe_sample():
    env = random.choice(ENVIRONMENTS)
    nodes, edges = create_base_graph(env, random.randint(4, 10))
    
    return {
        "environment": env,
        "event": "Safe",
        "nodes": nodes,
        "edges": edges
    }

def generate_unsafe_sample(threat_event):
    env = random.choice(ENVIRONMENTS)
    nodes, edges = create_base_graph(env, random.randint(3, 8))
    
    # Inject the threat
    threat_label = random.choice(THREAT_MAPPING[threat_event])
    threat_id = f"n{len(nodes)+1}"
    nodes.append({"id": threat_id, "label": threat_label, "attributes": []})
    
    # Inject the critical semantic edge that makes it unsafe
    person_id = "n1" # We always force a Person to be n1 in the base graph
    
    if threat_event in ["Gun Detection", "Knife Detection"]:
        critical_edge = "holding_or_wielding"
        edges.insert(0, {"source": person_id, "target": threat_id, "relationship": critical_edge})
        
    elif threat_event == "Fire Detection":
        critical_edge = "engulfing_or_spreading_to"
        random_target = random.choice(nodes)["id"]
        edges.insert(0, {"source": threat_id, "target": random_target, "relationship": critical_edge})
        # Add a secondary panic edge
        edges.insert(1, {"source": person_id, "target": threat_id, "relationship": "reacting_to_or_fleeing"})
        
    elif threat_event == "Fall Detection":
        critical_edge = "physically_interacting_with" # e.g. Person Falling -> Floor/Desk
        random_target = random.choice([n["id"] for n in nodes if n["label"] in ["Desk", "Chair", "Table", "Couch"]]) if any(n["label"] in ["Desk", "Chair", "Table", "Couch"] for n in nodes) else "n2"
        edges.insert(0, {"source": threat_id, "target": random_target, "relationship": critical_edge})

    return {
        "environment": env,
        "event": threat_event, # e.g., "Gun Detection"
        "nodes": nodes,
        "edges": edges
    }

# ==========================================
# 3. BUILD THE BALANCED DATASET
# ==========================================
def synthesize_dataset(output_filename, total_samples=2000):
    samples = []
    
    # 50% Safe
    num_safe = total_samples // 2
    for _ in range(num_safe):
        samples.append(generate_safe_sample())
        
    # 50% Unsafe (Equally distributed among the 4 threat types)
    unsafe_types = list(THREAT_MAPPING.keys())
    num_per_threat = (total_samples // 2) // len(unsafe_types)
    
    for threat in unsafe_types:
        for _ in range(num_per_threat):
            samples.append(generate_unsafe_sample(threat))
            
    # Shuffle so the model doesn't learn sequences
    random.shuffle(samples)
    
    # Write to JSONL
    with open(output_filename, 'w', encoding='utf-8') as f:
        for sample in samples:
            f.write(json.dumps(sample) + '\n')
            
    print(f"✅ Generated {total_samples} highly semantic scene graphs.")
    print(f"📊 Breakdown: {num_safe} Safe | {num_per_threat * len(unsafe_types)} Unsafe (Balanced across {len(unsafe_types)} classes)")
    print(f"📁 Saved to: {output_filename}")

# Run it!
if __name__ == "__main__":
    synthesize_dataset("./Data/json/master_train_data.jsonl", total_samples=2000)
```
